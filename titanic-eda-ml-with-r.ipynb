**split dataset**

```r
full.train <- full[1:891,]
full.test <- full[892:1309,]
```

**Modeling and Prediction**

```r
full.train$Survived <- as.factor(full.train$Survived)
```

**logistic Regrssion**

```r
log.mod <- glm(Survived ~ Pclass + Gender_encode + Gender_encode+ Age, family = binomial(link=logit), data=full.train)

summary(log.mod)
confint(log.mod)
```

```r
train.probs <- predict(log.mod, data=full.train, type="response")
table(full.train$Survived, train.probs > 0.5)

y_pred = ifelse(train.probs > 0.5, 1, 0)
y_pred

y_pred <- as.factor(y_pred)

str(y_pred)
class(full.train$Survived)
```

```r
library(ROCR)

confusionMatrix(y_pred, full.train$Survived)

fitpred = prediction(train.probs, full.train$Survived)
fitperf = performance(fitpred,"tpr","fpr")
plot(fitperf,col="green",lwd=2,main="ROC Curve")
abline(a=0,b=1,lwd=2,lty=2,col="gray")
```

**naiveBayes**

```r
library(e1071)

classifier <- naiveBayes(Survived ~ Pclass + Gender_encode + Age, data=full.train)
naive.pred <- predict(classifier, full.train, type="class") 
table(naive.pred, full.train$Survived) 


```

```r
confusionMatrix(naive.pred, full.train$Survived)
```

**randomForest**

```r
set.seed(675)

survived.formula <- as.formula("Survived ~ Pclass + Gender_encode + Age")
survive.model <- randomForest(formula = survived.formula, data = full.train, ntree = 500, mtry = 3, nodesize = 0.01 * nrow(full.test),  keep.forest = TRUE)
survive.model
```

```r
# Prediction (For prediction of survivals we will proceed with *Test* dataset)
Survived <- predict(survive.model, newdata = full.test)
Survived
```
